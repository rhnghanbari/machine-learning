```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import requests
import io
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler,scale
```

```python
url='https://raw.githubusercontent.com/Ali-Setareh/Machine-learning-in-physics-Project/main/Phase%201-Data%20Collection/Data/final%20data%20set.csv?token=ARIUNQAQAHT2VY6V7KPMWOTASOCH4'
s=requests.get(url, stream=True).content
meta=pd.read_csv(io.BytesIO(s))
```

```python
meta
```

```python
col_list=list(meta.columns)
col_list
```

make correct Data Frame and drop unessential columns

```python
Y=meta['Worldwide Gross ($)']
X=meta.iloc[:,13:292].join(meta.iloc[:,9])
X=X.drop(['User','Critic'],axis=1)
new_col_list=list(X.columns)
```

```python
X
```

min max scaler

```python
mm_scaler = MinMaxScaler()
mm_scaler.fit(X)
X_mmscaled = mm_scaler.transform(X)
X_df_mmscaled=pd.DataFrame(X_mmscaled)
X_df_mmscaled=X_df_mmscaled.set_axis(new_col_list, axis='columns')
```

```python
X_df_mmscaled
```

```python
X_df_mmscaled.to_csv('scaled_data.csv')
```

standard scaler

```python
s_scaler = StandardScaler()
X_sscaled = s_scaler.fit_transform(X) 
X_df_sscaled=pd.DataFrame(X_sscaled)
X_df_sscaled=X_df_sscaled.set_axis(new_col_list, axis='columns')
X_df_sscaled
```

```python
X_scaled = scale(X)
X_df_scaled=pd.DataFrame(X_scaled)
X_df_scaled=X_df_scaled.set_axis(new_col_list, axis='columns')
X_df_scaled
```
